```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import Perceptron
from sklearn.svm import SVC, LinearSVC
from sklearn.tree import DecisionTreeClassifier
```

```python
url = pd.read_csv('../data-extraction/url_collection.csv')
```

```python
url_df = pd.DataFrame(url)
url_df = url_df.sample(frac = 1)
display(url_df)
```

```python
url_arr = np.array(url_df)
print(url_arr)
```

```python
urls = [i[0] for i in url_arr]
```

```python
types = [i[1] for i in url_arr]
```

```python
tfidf_model = TfidfVectorizer().fit(urls)
sparse_result = tfidf_model.transform(urls)
```

```python
print(sparse_result)
```

```python
X_train, X_test, y_train, y_test = train_test_split(sparse_result, types, test_size=0.2, random_state=42)
```

```python
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, y_train)
score = random_forest.score(X_test, y_test)
print("score: {0:.2f} %".format(100 * score))
```

```python
mlp = MLPClassifier()
mlp.fit(X_train, y_train)
score = mlp.score(X_test, y_test)
print("score: {0:.2f} %".format(100 * score))
```

```python
lgr = LogisticRegression()
lgr.fit(X_train, y_train)
score = lgr.score(X_test, y_test)
print("score: {0:.2f} %".format(100 * score))
```

```python
perceptron = Perceptron()
perceptron.fit(X_train, y_train)
score = perceptron.score(X_test, y_test)
print("score: {0:.2f} %".format(100 * score))
```

```python
svc = SVC()
svc.fit(X_train, y_train)
score = svc.score(X_test, y_test)
print("score: {0:.2f} %".format(100 * score))
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
score = dt.score(X_test, y_test)
print("score: {0:.2f} %".format(100 * score))
```
